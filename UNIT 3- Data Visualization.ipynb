* few popular plotting libraries
1. matplotlib: low level,provides lots of freedom
2. pandas visualization: easy to use interface, built on              matplotlib
3. seaborn: high-level interface, great default styles
4. ggplot: based on R's ggplot2, uses grammar of graphics
5. plotly: can create interactive plots

```python
import matplotlib.pyplot as plt

a=[10,20,30,40,50,60,70]     #x-axis
b=[50,59,60,60,60,60,69]     #y-axis

plt.plot(a,b) #ploting the points

plt.xlabel('x-axis')
plt.ylabel('y-axis')  #labeling

plt.title('My 1st graph') #titling

plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax=plt.subplots() 

x=np.linspace(0,20,1000)   #(start point, end point, wave)
ax.plot(x,np.sin(x))
ax.axis('equal')

ax.annotate('local maximum', xy=(6.28,1),xytext=(10,4),arrowprops=dict(facecolor='green',shrink=0.05))
ax.annotate('local minimum', xy=(5*np.pi,-1),xytext=(4,-6),arrowprops=dict(arrowstyle='->',connectionstyle='angle3,angleA=0,angleB=-90'));

```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax=plt.subplots()#draw plot as fig

x=np.linspace(0,30,1000)
ax.plot(x,np.cos(x))
ax.axis('equal')

ax.annotate('local maximum', xy=(25,1), xytext=(10,4),
           arrowprops=dict(facecolor='pink',shrink=0.05))

ax.annotate('local minimun', xy=(5*np.pi, -1), xytext=(2,-6),
           arrowprops=dict(arrowstyle='->',facecolor='blue',
                          connectionstyle="angle,angleA=0,angleB=80"));

```

```python
import matplotlib.pyplot as plt

w=10
h=4
d=50

plt.figure(figsize=(w,h),dpi=d)
plt.axis([0,5,0,5])
x=[0,3,5]
y=[1,4,3.5]

label_x=1
label_y=4
arrow_x=3
arrow_y=4

arrow_properties=dict(facecolor='red',width=0.5,headwidth=8,shrink=2)

plt.annotate('maximum',xy=(arrow_x,arrow_y), xytext=(label_x,label_y), arrowprops=arrow_properties)

plt.plot(x,y)
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig=plt.figure()

ax=plt.axes(projection='3d')

z=np.linspace(0,2,200)
x=z*np.sin(25*z)
y=z*np.cos(25*z)

ax.plot3D(x,y,z,'green')
ax.set_title('3D plot')
plt.show()
```
